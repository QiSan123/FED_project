```python
import eikon as eikon
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import datetime as dt
from matplotlib.ticker import FuncFormatter
from functools import partial
```

```python
 eikon.set_app_key('1153655a813e49b09943821a11d273d879402478')
```

```python
rics = ['EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 
        'EDc9','EDc10','EDc11','EDc12',
        'EDc13','EDc14', 'EDc15','EDc16']
```

```python
table = eikon.get_timeseries (rics, fields ='CLOSE', start_date = '2003-07-01', end_date = '2004-06-30' , interval = 'daily', calendar='native')
```

```python
Dates = table.index.tolist()
EDc1 = (100-table['EDc1']).tolist()
EDc2 = (100-table['EDc2']).tolist()
EDc3 = (100-table['EDc3']).tolist()
EDc4 = (100-table['EDc4']).tolist()
EDc5 = (100-table['EDc5']).tolist()
EDc6 = (100-table['EDc6']).tolist()
EDc7 = (100-table['EDc7']).tolist()
EDc8 = (100-table['EDc8']).tolist()
EDc9 = (100-table['EDc9']).tolist()
EDc10 = (100-table['EDc10']).tolist()
EDc11 = (100- table['EDc11']).tolist()
EDc12 = (100-table['EDc12']).tolist()
EDc13 = (100-table['EDc13']).tolist()
EDc14 = (100-table['EDc14']).tolist()
EDc15 = (100-table['EDc15']).tolist()
EDc16 = (100-table['EDc16']).tolist()

table_r =pd.DataFrame({'Dates':Dates, 'EDc1': EDc1, 'EDc2': EDc2, 'EDc3': EDc3, 'EDc4': EDc4, 'EDc5': EDc5,
                 'EDc6': EDc6, 'EDc7': EDc7, 'EDc8': EDc8, 'EDc9': EDc9, 'EDc10': EDc10, 'EDc11': EDc11,
                 'EDc12': EDc12, 'EDc13': EDc13, 'EDc14': EDc14, 'EDc15': EDc15, 'EDc16': EDc16})
table_r = table_r[['Dates','EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 'EDc9','EDc10','EDc11','EDc12','EDc13','EDc14', 'EDc15','EDc16']]
table_r.fillna(method='ffill', inplace= True)
```

```python
# replacing outliers with mean rates
rates = ['EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 'EDc9','EDc10','EDc11','EDc12','EDc13','EDc14', 'EDc15','EDc16']

for x in rates:
    mean = sum(table_r[x])/ len(table_r[x])
    
    Q1 = table_r[x].quantile(0.25)
    Q3 = table_r[x].quantile(0.75)
    IQR = Q3 - Q1
    table_r[x] = table_r[x].mask((table_r[x]< (Q1 - 1.5* IQR)) | (table_r[x] > (Q3  + 1.5 * IQR)),mean) 
    table_r[x].fillna(method='ffill', inplace= True)
    table_r[x].fillna(method='bfill', inplace= True)
```

```python
table_r['EDc1_1D'] = (table_r['EDc1'] - table_r['EDc1'].shift(1))*100
table_r['EDc2_1D'] = (table_r['EDc2'] - table_r['EDc2'].shift(1))*100
table_r['EDc3_1D'] = (table_r['EDc3'] - table_r['EDc3'].shift(1))*100
table_r['EDc4_1D'] = (table_r['EDc4'] - table_r['EDc4'].shift(1))*100
table_r['EDc5_1D'] = (table_r['EDc5'] - table_r['EDc5'].shift(1))*100
table_r['EDc6_1D'] = (table_r['EDc6'] - table_r['EDc6'].shift(1))*100
table_r['EDc7_1D'] = (table_r['EDc7'] - table_r['EDc7'].shift(1))*100
table_r['EDc8_1D'] = (table_r['EDc8'] - table_r['EDc8'].shift(1))*100
table_r['EDc9_1D'] = (table_r['EDc9'] - table_r['EDc9'].shift(1))*100
table_r['EDc10_1D'] = (table_r['EDc10'] - table_r['EDc10'].shift(1))*100
table_r['EDc11_1D'] = (table_r['EDc11'] - table_r['EDc11'].shift(1))*100
table_r['EDc12_1D'] = (table_r['EDc12'] - table_r['EDc12'].shift(1))*100
table_r['EDc13_1D'] = (table_r['EDc13'] - table_r['EDc13'].shift(1))*100
table_r['EDc14_1D'] = (table_r['EDc14'] - table_r['EDc14'].shift(1))*100
table_r['EDc15_1D'] = (table_r['EDc15'] - table_r['EDc15'].shift(1))*100
table_r['EDc16_1D'] = (table_r['EDc16'] - table_r['EDc16'].shift(1))*100
table_1D = table_r.loc[:,['Dates', 'EDc1_1D', 'EDc2_1D', 'EDc3_1D', 'EDc4_1D', 'EDc5_1D', 'EDc6_1D',
                       'EDc7_1D', 'EDc8_1D', 'EDc9_1D','EDc10_1D','EDc11_1D','EDc12_1D','EDc13_1D',
                       'EDc14_1D', 'EDc15_1D','EDc16_1D']]
table_1D = table_1D[1:]
```

```python
daily = ['EDc1_1D', 'EDc2_1D', 'EDc3_1D', 'EDc4_1D', 
         'EDc5_1D', 'EDc6_1D','EDc7_1D', 'EDc8_1D', 
         'EDc9_1D','EDc10_1D','EDc11_1D','EDc12_1D',
         'EDc13_1D','EDc14_1D', 'EDc15_1D','EDc16_1D']

all_fut1D = []
for x in daily:
    all_fut1D.extend(table_1D[x])
```

```python
fig1, ax1 = plt.subplots(1,1, figsize=(7, 6))

ax1.hist(all_fut1D, bins = 30, range=[-15,15], normed = True, edgecolor='grey')


ax1.title.set_text('Distribution of Jul 2003 to Jun 2004 Daily Moves')

ax1.set_xlabel('bps')
ax1.set_xticks(np.arange(-20,21,2))

ax1.set_ylabel('Probability')
ax1.set_yticks(np.arange(0,0.25,0.02))

y_vals1 = ax1.get_yticks()
ax1.set_yticklabels(['{:3.2f}%'.format(x*100) for x in y_vals1])
```

```python
def count(list1,l,r):
    return len(list(x for x in list1 if l < x <= r))
```
