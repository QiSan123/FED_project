```python
rics = ['EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 
        'EDc9','EDc10','EDc11','EDc12',
        'EDc13','EDc14', 'EDc15','EDc16']
```

```python
table_1 = eikon.get_timeseries (rics, fields ='CLOSE', start_date = '2002-01-01', end_date = '2002-02-05' , interval = 'daily', calendar='native')
table_2 = eikon.get_timeseries (rics, fields ='CLOSE', start_date = '2002-01-01', end_date = '2002-10-31' , interval = 'daily', calendar='native')
table = pd.concat([table_1, table_2], ignore_index = False)
```

```python
table
```

```python
#changing the price to rates
Dates = table.index.tolist()
EDc1 = (100-table['EDc1']).tolist()
EDc2 = (100-table['EDc2']).tolist()
EDc3 = (100-table['EDc3']).tolist()
EDc4 = (100-table['EDc4']).tolist()
EDc5 = (100-table['EDc5']).tolist()
EDc6 = (100-table['EDc6']).tolist()
EDc7 = (100-table['EDc7']).tolist()
EDc8 = (100-table['EDc8']).tolist()
EDc9 = (100-table['EDc9']).tolist()
EDc10 = (100-table['EDc10']).tolist()
EDc11 = (100- table['EDc11']).tolist()
EDc12 = (100-table['EDc12']).tolist()
EDc13 = (100-table['EDc13']).tolist()
EDc14 = (100-table['EDc14']).tolist()
EDc15 = (100-table['EDc15']).tolist()
EDc16 = (100-table['EDc16']).tolist()

table_r =pd.DataFrame({'Dates':Dates, 'EDc1': EDc1, 'EDc2': EDc2, 'EDc3': EDc3, 'EDc4': EDc4, 'EDc5': EDc5,
                 'EDc6': EDc6, 'EDc7': EDc7, 'EDc8': EDc8, 'EDc9': EDc9, 'EDc10': EDc10, 'EDc11': EDc11,
                 'EDc12': EDc12, 'EDc13': EDc13, 'EDc14': EDc14, 'EDc15': EDc15, 'EDc16': EDc16})
table_r = table_r[['Dates','EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 'EDc9','EDc10','EDc11','EDc12','EDc13','EDc14', 'EDc15','EDc16']]
table_r.fillna(method='ffill', inplace= True)
```

```python
# replacing outliers with mean rates
rates = ['EDc1', 'EDc2', 'EDc3', 'EDc4', 'EDc5', 'EDc6','EDc7', 'EDc8', 'EDc9','EDc10','EDc11','EDc12','EDc13','EDc14', 'EDc15','EDc16']

for x in rates:
    mean = sum(table_r[x])/ len(table_r[x])
    
    Q1 = table_r[x].quantile(0.25)
    Q3 = table_r[x].quantile(0.75)
    IQR = Q3 - Q1
    table_r[x] = table_r[x].mask((table_r[x]< (Q1 - 1.5* IQR)) | (table_r[x] > (Q3  + 1.5 * IQR)),mean) 
    table_r[x].fillna(method='ffill', inplace= True)
    table_r[x].fillna(method='bfill', inplace= True)
```
